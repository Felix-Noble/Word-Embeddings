```python
### KMEANS clustering and sillhouette score analysis ###
# Initialisation 
km_args = {"n_clusters":5, "random_state":2025, "n_init":100} # initialise km_args dictionary 
from sklearn.metrics import silhouette_score # import scoring function 
```

```python
# Running
scores = []
# data = pipe.data 
data = umap # use [umap] data for analysis 
start_nc=2;end_nc=20 # start n. clusters and end n. clusters
for n_c in range(start_nc,end_nc+1):
    km_args["n_clusters"] = n_c # change n_clusters arg for KMEANS method 
    kmeans = pipe.KMEANS(data, km_args) # get cluster labels
    scores.append(silhouette_score(data, kmeans.values.ravel())) # store sillhouette score in list 
```

```python
# Sillhouette analysis visualisation

plt.plot(range(start_nc,len(scores)+start_nc), np.array(scores).reshape(-1));plt.title("N. Clusters over Silhouette Score"); plt.xlabel("N. Clusters")
plt.ylabel("Silhouette Score");plt.xticks(range(2,len(scores)+2))
# calculate highest score 
high = max(scores) 
# plot vertical line at highest score
plt.axvline(scores.index(high)+start_nc, c="red", linestyle="--"); plt.legend(["Score", f"Max: {scores.index(high)+2}"]) 
# plt.savefig(os.path.join(pipe.output, f"umap_sort KMeans S_score {start_nc}-{end_nc}.png"), dpi=300)
```

```python
# Clustering Solution Visualisation

km_args = {"n_clusters":2, "random_state":2025, "n_init":100} # args for kmeans clustering (n_clusters arbitrary)
kmeans = pipe.KMEANS(umap, km_args) # get kmeans labels

labels = sorted(set(kmeans.values.ravel())) # identify and sort unique label values (length will always be n_clusters)
plt.figure(figsize=(10,10)) # initialise figure 
for label, color in zip(labels, cols): # iterate over labels and colours simultaneously (with zip)
    idx = kmeans.values.ravel() == label # find the datapoints belonging to cluster 
    plt.scatter(umap.iloc[idx, 0], umap.iloc[idx, 1], c=color, label=f"Cl:{label + 1} N={np.count_nonzero(idx)}") # scatter cluster data points

plt.xlim(min(umap.iloc[:,0])-1, max(umap.iloc[:,0])+1); plt.ylim(min(umap.iloc[:,1])-1, max(umap.iloc[:,1])+1) # extend x/ylim for better viewing
plt.legend(loc="upper right") # add the legend (determined by label argument from plt.scatter())
plt.xlabel(f"Dim {1}");plt.ylabel(f"Dim {2}") # x and y labels
plt.title(f"Spotify_playlists UMAP 2 features KMeans {km_args['n_clusters']} clusters") # title
# plt.savefig(pipe.output / f"{km_args['n_clusters']} clusters coloured.png", dpi=300) # add filename to pipe.output dir for saving
plt.show() # show figure after saving 
```

```python
### FUZZY C MEANS clustering ###
# Initialisation 
fuzzyc_args = {"n_clusters": 4, "error":0.005, "maxiter": 1000, "random_state":None} # initialise args dictionary 
```

```python
# Running
scores = []
stds = []
n_sample = 50 # number of samples to draw (Fuzzy C Means is stochastic)
for n_c in range(2,10):
    temp = np.zeros(n_sample) # preallocate array for score values 
    for i in range(n_sample):
        fuzzyc_args["n_clusters"] = n_c # change n. clusters argument 
        kfuzzy, score = pipe.FUZZYCMEANS(umap, fuzzyc_args, score=True)
        temp[i] = score # store value at specific index of temp array 
    scores.append(np.mean(temp)) # store mean
    stds.append(np.std(temp)) # store std

plt.errorbar(range(2,len(scores)+2), np.array(scores).reshape(-1), yerr=stds, fmt='--') # plot data as errorbar
plt.title("Fuzzy C Means score")
high = max(scores) # identify highest scoreing solution
plt.axvline(scores.index(high)+2, c="red", linestyle="--");plt.legend(["scores", "max"])
#plt.save(pipe.output / "YOUR FILENAME HERE.png", dpi=300)
plt.show() # show figure after saving 
```

```python
# Visualise selected clustering solution
fuzzyc_args["n_clusters"] = 5
kfuzzy, score = pipe.FUZZYCMEANS(umap, fuzzyc_args, score=True)

def plot_kfuzzy(axes, fuzzy_labels, label_i):
    for i in range(umap.shape[0]):
        a = (fuzzy_labels[i,label_i])/np.max(fuzzy_labels) # determine proportional alpha (transparency) value 
        axes.scatter(umap.iloc[i,0], umap.iloc[i,1], alpha=a, c=cols[label_i])  # scatter indiviudal point on pre indexed axes

```

```python
n_rows = 2 # n rows = round down after dividing total by 2
n_cols = math.ceil(kfuzzy.shape[1]/n_rows) # n cols = round up after dividing total by 2
sub_fig_width = 4;sub_fig_height = 4
fig, axes = plt.subplots(n_rows, n_cols, figsize=(sub_fig_width*n_cols, sub_fig_height*n_rows)) # initialise subplot figure 
axes = axes.flatten() # flatten axes for 1D indexing ([i] instead of [x,y]): Requires axes.size > 1

for label_i in range(0,n_rows * n_cols):
    ax = axes[label_i] # set which axes to plot on
    if label_i > kfuzzy.shape[1]-1:
        ax.set_title(f"EMPTY", fontsize=12) # make sure that empty subplots are demarkated 
        continue # ensure loop won't try to access non_existent kfuzzy rows (continue skips to next loop start)
    plot_kfuzzy(ax, kfuzzy, label_i) # call plotting function 
    
    ax.set_title(f"Cluster {label_i+1}", fontsize=12) # set title and x/y labels
    ax.set_xlabel("UMAP 1")
    ax.set_ylabel("UMAP 2")

#plt.save(pipe.output / "YOUR FILENAME HERE.png", dpi=300)
plt.show() # show figure after saving
```
